```python
import numpy as np
def score(y_pred, y_true):
    return np.sum(y_pred==y_true)/y_pred.shape[0]
```

```python
from kr_means import *
import mglearn
import numpy as np
import pandas as pd
from sofa16qamlib import *

num_sym = 500
sym_tx = np.random.choice(np.arange(0,16), size=num_sym) #Generamos indices de símbolos aleatorios (0,15)
sig_tx=modulate(sym_tx) #modula

Phi = 10 #ruido de fase
Rot=0 #rotación
SNR=30
errv=[]

sig_rx=noiselnl(sig_tx,SNR,Phi,Rot) #agrega ruido
X = pd.DataFrame(np.asarray([[sig_rx[i].real, sig_rx[i].imag] for i in range(len(sig_rx))]))

assignments_list = []

for b in range(0,5):
    b_normalized = b/10
    kmeans = KRMeans(n_clusters=16, verbose=False, alpha=1, beta=b_normalized)
    kmeans.fit(X)
    
    labels = np.asarray(demodulate(kmeans.cluster_centers_.iloc[:,0]+1j*kmeans.cluster_centers_.iloc[:,1]))           #demodula
    assignments = labels[kmeans.labels_]
    assignments_list.append(assignments)

    errores=biterr(assignments,sym_tx)
    BER_tra=(errores/(num_sym*4))
    print('Error de bits con beta',b_normalized,'es',BER_tra)
    print('Error con beta',b_normalized,'es',1-score(assignments,sym_tx))
    %store assignments_list
```

```python
mglearn.discrete_scatter(X.iloc[:, 0], X.iloc[:, 1], assignments_list[0])
```
